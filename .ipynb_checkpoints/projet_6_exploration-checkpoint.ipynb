# 1 Pr√©paration du data set

Filtrer uniquement les b√¢timents non r√©sidentiels : on utilise la colonne 'BuildingType' en faisant afficher les valeurs qu'elle peut prendre :

```python
building_consumption['BuildingType'].value_counts()
```

et on filtre les batiments non r√©sidentiels : 

```python
non_residential_types = ['NonResidential', 'Nonresidential COS', 'Nonresidential WA', 'Campus']
df_filtered = building_consumption[building_consumption['BuildingType'].isin(non_residential_types)].copy()
```

On filtre pour ne garder les batiments > 20000 square feet  

*‚ÄúSeattle‚Äôs Building Energy Benchmarking Program (SMC 22.920) requires owners of non-residential and multifamily buildings greater than 20,000 square feet to track energy performance and annually report to the City of Seattle.‚Äù*  

Ce qu'on est cens√© avoir en th√©orie, mais cela va d√©j√† supprimer des anomalies √©videntes, am√©liorer la coh√©rence et r√©duire le bruit 

```python
print("Avant filtrage GFA ‚â• 20 000 :", len(df_filtered))
df_filtered = df_filtered[
    pd.to_numeric(df_filtered["PropertyGFATotal"], errors="coerce") >= 20000
].copy()
print("Apr√®s filtrage GFA ‚â• 20 000 :", len(df_filtered))
```

Nous allons regarder de plus pres la colonne outliers

```python
print(df_filtered[df_filtered['Outlier'].notna()])
```

Avant de supprimer les colonnes qui ne nous seront pas utiles pour l'√©tude, on cr√©e nos variables d√©riv√©es BuildingAge et ParkingRatio :

```python
# BuildingAge : √¢ge du b√¢timent au moment de la campagne 2016
df_filtered['BuildingAge'] = 2016 - df_filtered['YearBuilt']

# ParkingRatio : ratio parking / surface totale
df_filtered['ParkingRatio'] = df_filtered['PropertyGFAParking'] / df_filtered['PropertyGFATotal']

# V√©rifier les nouvelles colonnes
df_filtered[['YearBuilt', 'BuildingAge', 'PropertyGFATotal', 'PropertyGFAParking', 'ParkingRatio']].head()
```

### 2 Variables cibles 

```python
target_energy = 'SiteEnergyUse(kBtu)'
target_ghg = 'TotalGHGEmissions'
```

V√©rifier les valeurs manquantes des targets :

```python
print("Valeurs manquantes :")
print(df_filtered[[target_energy, target_ghg]].isnull().sum())
```

Supprimer lignes avec target manquante : 

```python
df_filtered.dropna(subset=[target_energy, target_ghg], inplace=True)
print("Nombre de lignes apr√®s suppression valeurs manquantes targets :", df_filtered.shape[0])
```

```python
features = [
    'PropertyGFATotal', 'PropertyGFABuilding(s)', 'PropertyGFAParking',
    'NumberofBuildings', 'NumberofFloors', 'BuildingAge',
    'PrimaryPropertyType', 'LargestPropertyUseType',
    'CouncilDistrictCode', 'Neighborhood', 'Latitude', 'Longitude',
    'ParkingRatio'
]
```

```python
df_model = df_filtered[features + [target_energy, target_ghg]].copy()
```

### 4 Statistiques descriptives

```python
print("\nStatistiques descriptives (num√©riques) :")
display(df_model.describe())
```

Colonnes cat√©gorielles

```python
categorical_cols = ['PrimaryPropertyType', 'LargestPropertyUseType', 'Neighborhood']
for col in categorical_cols:
    print(f"\nDistribution de {col}:")
    print(df_model[col].value_counts())
```

### 5 Identification des outliers

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    print(f"{column} : {len(outliers)} valeurs aberrantes ({len(outliers)/len(df)*100:.2f}%)")
    return outliers

num_cols = ['SiteEnergyUse(kBtu)', 'TotalGHGEmissions', 'PropertyGFATotal', 
            'PropertyGFABuilding(s)', 'PropertyGFAParking', 'ParkingRatio']

for col in num_cols:
    detect_outliers_iqr(df_model, col)

# Log-transform pour r√©duire l'effet des outliers
for col in ['SiteEnergyUse(kBtu)', 'TotalGHGEmissions', 'PropertyGFATotal', 'PropertyGFABuilding(s)', 'PropertyGFAParking']:
    df_model[col + '_log'] = np.log1p(df_model[col])
```

### 6 Visualisation

Histogrammes + densit√©

```python
for col in ['SiteEnergyUse(kBtu)_log', 'TotalGHGEmissions_log', 
            'PropertyGFATotal_log', 'PropertyGFABuilding(s)_log', 'ParkingRatio']:
    plt.figure()
    sns.histplot(df_model[col], kde=True, bins=40, color='skyblue')
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel("Fr√©quence")
    plt.show()
```

Variables cat√©gorielles : consommation par type de b√¢timent

Ces histogrammes montrent l‚Äôeffet b√©n√©fique de la transformation logarithmique sur les variables continues fortement asym√©triques.
La consommation d‚Äô√©nergie et les √©missions de CO‚ÇÇ deviennent plus sym√©triques, facilitant leur mod√©lisation.
La surface totale suit une distribution plus √©quilibr√©e apr√®s log, tandis que la part de parking reste tr√®s concentr√©e.
On observe √©galement que certaines variables (PropertyGFATotal et PropertyGFABuildings) sont quasi √©quivalentes et pourront √™tre simplifi√©es dans la phase de s√©lection de features.

```python
for col in ['PrimaryPropertyType', 'LargestPropertyUseType']:
    plt.figure(figsize=(12,6))
    sns.boxplot(x=col, y='SiteEnergyUse(kBtu)_log', data=df_model)
    plt.xticks(rotation=90)
    plt.title(f"Consommation √©nerg√©tique par {col} (log-transform)")
    plt.show()
```

Quantitative vs quantitative : scatter plots log-log

Les boxplots montrent une forte variabilit√© de la consommation d‚Äô√©nergie selon le type de b√¢timent.  
Les h√¥pitaux, laboratoires et data centers affichent des niveaux de consommation log-transf√©r√©e parmi les plus √©lev√©s,  
alors que les bureaux, √©coles et commerces sont beaucoup plus sobres.

La dispersion interne aux cat√©gories (hauteur des bo√Ætes) sugg√®re que la taille et d‚Äôautres facteurs structurels  
influent aussi sur la consommation, mais l‚Äôusage reste la variable qualitative la plus d√©terminante.

`LargestPropertyUseType` apporte une vision plus fine que `PrimaryPropertyType`,  
car il refl√®te l‚Äôusage dominant r√©el du b√¢timent. Ces deux variables seront retenues comme features importantes.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='PropertyGFATotal_log', y='SiteEnergyUse(kBtu)_log', data=df_model)
plt.title('Consommation d‚Äô√©nergie vs Surface totale (log-log)')
plt.xlabel('PropertyGFATotal (log scale)')
plt.ylabel('SiteEnergyUse(kBtu) (log scale)')
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='PropertyGFABuilding(s)_log', y='TotalGHGEmissions_log', data=df_model)
plt.title('√âmissions GES vs Surface b√¢timent (log-log)')
plt.xlabel('PropertyGFABuilding(s) (log scale)')
plt.ylabel('TotalGHGEmissions (log scale)')
plt.show()
```

S√©lection des colonnes num√©riques seulement

Les graphiques ci-dessus montrent des relations log-log lin√©aires entre la surface des b√¢timents et leur consommation d‚Äô√©nergie, 
ainsi qu‚Äôentre la surface b√¢tie et les √©missions de gaz √† effet de serre.

Ces relations traduisent un effet d‚Äô√©chelle : plus un b√¢timent est grand, plus il consomme et √©met, 
mais de mani√®re sous-proportionnelle, sugg√©rant certaines √©conomies d‚Äô√©chelle.

La corr√©lation tr√®s forte entre les deux paires de variables confirme que la taille du b√¢timent est un facteur cl√©, 
et que les √©missions de GES d√©coulent directement de la consommation √©nerg√©tique.
Les transformations logarithmiques rendent ces relations plus lisibles et facilitent la mod√©lisation.

```python
numeric_cols = df_model.select_dtypes(include=np.number)
```

Matrice de corr√©lation

```python
plt.figure(figsize=(12,10))
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corr√©lation (colonnes num√©riques)")
plt.show()
```

- La matrice montre une **forte corr√©lation (r ‚âà 0.95)** entre `PropertyGFATotal` et `PropertyGFABuildings`,
  ce qui indique une redondance entre ces deux variables (elles d√©crivent toutes deux la surface du b√¢timent).
- La **consommation totale d‚Äô√©nergie** (`SiteEnergyUseWN(kBtu)`) est positivement corr√©l√©e
  avec la surface (`PropertyGFATotal`) et le nombre d‚Äô√©tages (`NumberofFloors`),
  ce qui est coh√©rent avec les attentes physiques.
- L‚Äô**√¢ge du b√¢timent** (`Age`) pr√©sente une corr√©lation mod√©r√©e avec la consommation,
  sugg√©rant un effet plus limit√© de l‚Äôanciennet√© sur la performance √©nerg√©tique.
- Les ratios comme `ParkingShare` montrent peu de corr√©lation,
  ce qui indique qu‚Äôils ne sont pas des d√©terminants majeurs de la consommation.
- Aucune suppression n‚Äôest effectu√©e √† ce stade : cette analyse est uniquement exploratoire.
  Les corr√©lations fortes seront **prises en compte lors de la s√©lection des features** dans la mod√©lisation.
